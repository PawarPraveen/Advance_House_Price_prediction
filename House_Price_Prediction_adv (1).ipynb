```python
cat_cols = df.select_dtypes(include='object').columns
print(cat_cols)
```

```python
cardinality = df[cat_cols].nunique()
```

```python
low_cardinality = [col for col in cat_cols if df[col].nunique() <= 10]
high_cardinality = [col for col in cat_cols if df[col].nunique() > 10]

print("Low cardinality feature are", low_cardinality)
print("high cardinality feature are", high_cardinality)
```

```python
from sklearn.preprocessing import LabelEncoder
df_low = pd.get_dummies(df[low_cardinality])
df_high = df[high_cardinality].apply(LabelEncoder().fit_transform)

numerical_col = df.select_dtypes(exclude = 'object').columns
df_final = pd.concat([df[numerical_col], df_low, df_high], axis = 1)
```

```python
import xgboost
import shap
x = df_final.drop(columns = 'SalePrice')
y = df_final['SalePrice']

model = xgboost.XGBRegressor()
model.fit(x,y)

explainer = shap.Explainer(model)
shap_value = explainer(x)

shap.summary_plot(shap_value,x)
```

```python
import numpy as np
import pandas as pd

mean_shap = np.abs(shap_value.values).mean(axis = 0)

shap_importance = pd.DataFrame({
     'Feature' : x.columns,
     'MeanAbsShap' : mean_shap
}).sort_values(by = 'MeanAbsShap', ascending = True )

threshold = 0.001
low_impact = shap_importance[shap_importance['MeanAbsShap'] < threshold]['Feature'].tolist()
print("Feature to drop :", low_impact)
```

```python
num_columns_before = x.shape[1]
x_reduced = x.drop(columns=low_impact)
num_columns_after = x_reduced.shape[1]
num_columns_dropped = num_columns_before - num_columns_after

# Step 5: Output
print("Features to drop:", low_impact)
print(f"Total columns dropped: {num_columns_dropped}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer


# Step 3: Impute missing values
imputer = SimpleImputer(strategy='mean')
x_reduced_imputed = pd.DataFrame(imputer.fit_transform(x_reduced), columns=x_reduced.columns)

# Step 4: Train/test split
X_train, X_test, y_train, y_test = train_test_split(x_reduced_imputed, y, test_size=0.2, random_state=42)

# Step 5: Train Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Step 6: Predict and evaluate
y_pred = lr_model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"\nLinear Regression RMSE: {rmse:.4f}")
print(f"\n Linear Regression R² Score: {r2:.4f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import shap

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Separate numeric and categorical columns
numeric_cols = x.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = x.select_dtypes(include=['object', 'category']).columns

# Numeric pipeline
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean'))
])

# Categorical pipeline
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

# Combine preprocessors
preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_cols),
    ('cat', categorical_transformer, categorical_cols)
])

# Linear Regression pipeline
lr_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', LinearRegression())
])

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Train the model
lr_pipeline.fit(X_train, y_train)

# Predict
y_pred_lr = lr_pipeline.predict(X_test)

# Evaluate
rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred_lr))
r2_lr = r2_score(y_test, y_pred_lr)
print(f"✅ Linear Regression RMSE: {rmse_lr:.2f}")
print(f"✅ Linear Regression R² Score: {r2_lr:.4f}")

# Visualize actual vs. predicted
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred_lr, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel("Actual House Price")
plt.ylabel("Predicted House Price")
plt.title("Linear Regression: Actual vs. Predicted Prices")
plt.grid(True)
plt.tight_layout()
plt.show()

# SHAP Interpretation
# Get preprocessed training data
X_train_transformed = preprocessor.fit_transform(X_train)
feature_names = preprocessor.get_feature_names_out()

# Fit model separately for SHAP
lr_model = LinearRegression()
lr_model.fit(X_train_transformed, y_train)

# Use SHAP's LinearExplainer
explainer = shap.LinearExplainer(lr_model, X_train_transformed, feature_names=feature_names)
shap_values = explainer(X_train_transformed)

# SHAP Summary Plot
shap.summary_plot(shap_values.values, X_train_transformed, feature_names=feature_names)
```
